```python
access_list_copy['A/P Invoice'] = access_list_copy['A/P Invoice'].map({'Full Authorization' : 0, 'No Authorization' : 1, 'Read-Only' :2})
```

```python
access_list_copy
```

```python
access_list_copy['Adding Customer/Vendor Master Data'] = access_list_copy['Adding Customer/Vendor Master Data'].map({'Full Authorization' : 0, 'No Authorization' : 1, 'Read-Only' :2})
```

```python
access_list_copy['Adding Lead BP'] = access_list_copy['Adding Lead BP'].map({'Full Authorization' : 0, 'No Authorization' : 1, 'Read-Only' :2})
```

```python
access_list_copy['Remove Business Partner'] = access_list_copy['Remove Business Partner'].map({'Full Authorization' : 0, 'No Authorization' : 1, 'Read-Only' :2})
```

```python
access_list_copy
```

```python
auth_ls = []
conflict_ls = []
for ind, i in access_list_copy.iterrows():
    conf_id_str = ''
    auth_bool = 'Yes'
    if(i['A/P Invoice'] == 0 or i['A/P Invoice'] == 2):
        if(i['Adding Customer/Vendor Master Data'] == 0):
            conf_id_str += 'CRID001 '
            auth_bool = 'No'
        if(i['Adding Lead BP'] == 0):
            conf_id_str += 'CRID002 '
            auth_bool = 'No'
        if(i['Remove Business Partner'] == 0):
            conf_id_str += 'CRID003'
            auth_bool = 'No'
    conflict_ls.append(conf_id_str)
    auth_ls.append(auth_bool)
```

```python
for i in range(len(conflict_ls)):
    conflict_ls[i] = conflict_ls[i].replace(' ', ',')
```

```python
conflict_ls
```

```python
access_list_copy['Authorised Yes/No'] = auth_ls
```

```python
access_list_copy['Conflicting rules'] = conflict_ls
```

```python
access_list_copy
```

```python
invoices_copy
```

```python
auth_ls_invoices = []
conf_ls_invoices = []
for ind, i in invoices_copy.iterrows():
    auth_ls_invoices.append(access_list_copy[access_list_copy['User'] == i['User']]['Authorised Yes/No'].tolist())
    conf_ls_invoices.append(access_list_copy[access_list_copy['User'] == i['User']]['Conflicting rules'].tolist())
```
